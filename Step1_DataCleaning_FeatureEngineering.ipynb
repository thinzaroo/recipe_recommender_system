```python
import pandas as pd
import numpy as np
import json
```

```python
#dataset will be a python dictionary.
with open('data/recipes_full.json') as f:
  dataset = json.load(f)

#dataFrame will be a Pandas dataframe.
dataFrame = pd.read_json('data/recipes_full.json')
```

# Feature Engineering

Goals:

- convert JSON to csv
- Feature engineer new columns

```python
dataFrame.head(3)
```

```python
# separate "nutr_per_ingredient" column to 6 separate columns
df1 = pd.DataFrame(dataFrame['nutr_values_per100g'].values.tolist())
df1.columns = 'nutr_values_per100g_'+ df1.columns
df1.head(5)
```

```python
# 1. Append the 6 new columns to dataFrame
df = pd.concat([dataFrame, df1],axis=1)
df.head(3)
```

```python
# 2. Extract unique domain from the url and add to the dataFrame
# url = http://cookeatshare.com/recipes/salt-free-low-cholesterol-sugar-cookies-6256
# new column: domain = cookeatshare.com
import urllib

df['protocol'],df['url_domain'],df['path'],df['query'],df['fragment'] = zip(*[urllib.parse.urlsplit(x) for x in df['url']])

# drop the rest except domain
df = df.drop(['protocol', 'path', 'query', 'fragment'], axis=1)

df.head(3)
```

```python
# list all unique domains
print('Total unique domains count:',len(df['url_domain'].unique()))
df['url_domain'].unique()
```

```python
# 3. Get total ingredients count, and add column to dataFrame
df['ingredients_count'] = df.apply(lambda row: len(row.ingredients), axis=1)
df.head(3)
```

```python
# 4. Get instructions steps count, and add to dataFrame
df['instructions_steps'] = df.apply(lambda row: len(row.instructions) , axis=1)
df.head(3)
```

```python
# 5. simplify the ingredients column, only take the first comma-separated item

def simplify_ingredients(row):
    arr_simple_ingredients = []
    for x in row:
        arr_simple_ingredients.append(list(x.values())[0].split(',')[0])
    return arr_simple_ingredients
    
df['ingredients_simple'] = df.apply(lambda row: simplify_ingredients(row.ingredients) , axis=1)
df.head(3)
```

```python
df['instructions_simple'] = df.apply(lambda row: simplify_ingredients(row.instructions) , axis=1)
df.head(3)
```

```python
# Drop the unused columns
columns_to_drop = ['fsa_lights_per100g', 'ingredients', 'instructions' , 'nutr_per_ingredient',
                   'nutr_values_per100g', 'unit','quantity','weight_per_ingr']
df = df.drop(columns = columns_to_drop).dropna()
df.head(3)
```

```python
# save to csv
df.to_csv('data/recipes_feature_engineered.csv', index=False)
```

```python
# separate train/test files

import pandas as pd
df = pd.read_csv('data/recipes_feature_engineered.csv')
df_test = df[df['partition'] == 'test'].copy()
df_train = df[df['partition'] != 'test'].copy()

df_train.to_csv('data/recipes_feature_engineered_train.csv', index=False)
df_test.to_csv('data/recipes_feature_engineered_test.csv', index=False)
```
